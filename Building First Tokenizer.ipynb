```python
# Filter only sanskrit chars
import re

def remove_non_sanskrit(text):
    # Devanagari Unicode block: U+0900тАУU+097F
    return re.sub(r'[^\u0900-\u097F\s]', '', text)

text = remove_non_sanskrit(text)
```

```python
# here are all the sanskrit characters that occur in this text
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(''.join(chars))
print(vocab_size)
```

## Temporary Encoder and Decoder, by character Emumerization

```python
# create a mapping from characters to integers
stoi = { ch:i for i,ch in enumerate(chars) }
itos = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
decode = lambda l: ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string

print(encode("рдирдорд╕реНрддреЗ"))
print(decode(encode("рдирдорд╕реНрддреЗ")))
```

## Real Encoder

```python
# Sanskrit chars belong to an upper range, first they're converted to UTF-8 bytes, then bytes are down-converted to 0.255 range,
# i.e. raw bytes is just a display format, we're converting display format of UTF-8 to int and bringing that down.
tokens = text.encode("utf-8") # tokens are raw bytes
tokens = list(map(int, tokens)) # tokens are convert to a list of integers in range 0..255 for convenience
print('---')
print(text[:1000])
print("length:", len(text))
print('---')
print(tokens[:1000]) # for sanskrit, each char == 3 tokens
print("all tokens length:", len(tokens))
print("input tokens length:", len(tokens)//3)
```

```python
# These are the bytes pair that occur most frequently, doesn't mean the actual token pair, it's byte pairs
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]): # Pythonic way to iterate consecutive elements
        counts[pair] = counts.get(pair, 0) + 1
    return counts

stats = get_stats(tokens)
# print(stats)
print(sorted(((v,k) for k,v in stats.items()), reverse=True))
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge(ids, pair, idx):
  newids = []
  i = 0
  while i < len(ids):
    if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
      newids.append(idx)
      i += 2
    else:
      newids.append(ids[i])
      i += 1
  return newids
```

## Test Run the BPE

```python
# ---
orig_vocab_size = 256
desired_vocab_size = 260 # the desired final vocabulary size
num_merges = desired_vocab_size - orig_vocab_size
ids = list(tokens) # copy so we don't destroy the original list

merges = {} # (int, int) -> int
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key=stats.get)
  idx = orig_vocab_size + i
  print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx
```

```python
# ---
orig_vocab_size = 256
desired_vocab_size = 5000 # the desired final vocabulary size
num_merges = desired_vocab_size - orig_vocab_size
ids = list(tokens) # copy so we don't destroy the original list

merges = {} # (int, int) -> int
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key=stats.get)
  idx = orig_vocab_size + i
  print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx
  if idx % 10 == 0:
      print(f"compression ratio: {len(tokens) / len(ids):.2f}X")
```

## Save the Tokenizer

```python
import json

# After training with orig_vocab_size=128
tokenizer_config = {
    "merges": [[list(pair), idx] for pair, idx in merges.items()],  # Convert tuple keys to lists
    "orig_vocab_size": 256
}

with open("sanskrit_bpe.json", "w") as f:
    json.dump(tokenizer_config, f, indent=2)
```
